### Preprocessing and data cleaning

We'll be preparing the dataset to be used later on for feature engineering and model training. 

```python
to_keep = ['cancellation_policy', 'bed_type', 'room_type', 'neighbourhood_cleansed', 'city', 'host_response_time', 'property_type']

# These columns are specifically columns of `object` datatype that will be of use to us. We can safely remove the others. They are all categorical. 
# Since the dataset only has listings from the US, we can take the liberty to remove the `country` and `country_code` columns.

data_num = data.select_dtypes(include=['int64', 'float64'])

# Merging data_num and data[to_keep] into a new dataframe called data_new. 
# Here, we are removing the columns that are not of much use to us, like `summary`, `description`, and such.

# As for the column `host_location`, it can be removed since the `city` column gives us the same information, minus the country name. 
# The same goes for `smart_location` - it is covered in the `city` column. 
# Same for `neighbourhood` - it is covered in the `neighbourhood_cleansed` column.
# Same for `street` - covered in `neighbourhood_cleansed` and `city`.

# `neighbourhood_cleansed` better represents the neighbourhood of the host, since it is a cleaned version of the `neighbourhood` column. There 
# are also much fewer null values in this column. 

data_new = pd.concat([data_num, data[to_keep]], axis=1)

data_new.head()
```

```python
df_new = data_new.copy()
```

```python
data_new.describe()
```

```python
data_new.columns
```

```python
# Manually observing the dataset and removing additional columns that aren't of any use. 

data_new.drop(['picture_url', 'id', 'scrape_id', 'host_id', 'picture_url.width', 'picture_url.height'], inplace=True, axis=1)
```

```python
data_new.describe()
```

```python
# Removing the columns 'geolocation.lon' and 'geolocation.lat' since they are duplicates of the 'longitude' and 'latitude' columns

data_new.drop(['geolocation.lon', 'geolocation.lat'], inplace=True, axis=1)
```

```python
data_new.columns
```

```python
data_new.shape
```

```python
# Printing the number of nan or null values in each column

data_new.isnull().sum()
```

```python
# We see here that a few columns like `square_feet`, `weekly_price`, `monthly_price` have too many missing values to be ignored. 
# Dropping these columns is hence the best option.

# But we also see columns like `host_response_time` and `review_scores_rating` also have many missing values (around 25% of the rows). 
# Dropping these columns would mean losing a lot of data. However, considering the fact that about 75% of the data would still be
# available, we can drop only those rows that have null values in these columns, and not drop the whole column.

# These are also important features that are required while determining the price of a listing.  

# This is a decision made after considering that about 100,000 listings would still be available for analysis, and that 
# would still be a very good number to work with (~ 75% of the dataset). 
```

```python
# If the number of null values of a column is greater than 25% of the total number of rows, then we 
# drop that column

for col in data_new.columns:
    
    if (data_new[col].isnull().sum() > 0.25 * data_new.shape[0]).any():
        data_new.drop(col, inplace=True, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Encoding all columns of `object` datatype. This is the same as using OrdinalEncoder(), but we use LabelEncoder() here since we are dealing with 
# the columns one by one, with each having shape (n_samples, )

# We avoid OneHotEncoding() since it would result in a very large number of columns (eg. `neighbourhood_cleansed` alone has about 1200 unique 
# values), and would be computationally very expensive. 

for col in to_keep:
    try:
        data_new[col] = le.fit_transform(data_new[col])
    except KeyError:
        pass
```

### Feature Engineering

Not all features from here might contribute much to the model. Hence, the best subset of them will be discovered, and then used for the model. 

```python
# Examining the correlation of the features with the target variable `price`

data_new.corr(numeric_only=True)['price'].sort_values(ascending=False).drop('price')
```

```python
# Straight away, we see there are a few columns that have a very low correlation with the target variable `price`. Except for the 
# first 9 columns, the rest have a correlation of less than 0.1. If we go a further 3 features or so, it goes less than 0.05 (5%). 

# Right from this, it is quite clear that the neighbourhood, or even the city of the listing is a very trivial factor in 
# determining the price of the listing - which is quite surprising, considering the general fact that rents might change with location, in 
# addition to the number of beds, bathrooms, accomodates, etc. This doesn't seem to be the case with Airbnb listings. 

# We'll later apply PCA to see if we can reduce the number of features in the dataset and use only the most important ones. 
```

```python
data_new.isnull().sum()
```

```python
# Preparing various input datasets for experiments:

# 1) Regular dataset: X
# 2) Standardised dataset: X_std
# 3) PCA Reduced dataset: X_pca
```

```python
# 1) Regular Dataset

X = data_new.drop('price', axis=1)  # Stores the input attributes data
y = data_new['price']                 # Stores the target variable data 

# # Import Scikit-Learn's train_test_split to split the data into 70% training and 30% testing. `random_state = 42` for reproducibility. 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)

# Inspecting the shape of the dataset and taking a peak into it

print(f"\nNumber of samples and attributes of X: {X.shape}\n")
X.head()
```

```python
# 2) Standardized Dataset

# Standardizing the data subtracts the mean and scales the values to unit standard deviation
# It is applied to the input data to bring values to a standard range.

# Using the StandardScaler class to transform the regular data to Standardized data

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_std = pd.DataFrame(sc.fit_transform(X))

# Using train_test_split to create a new split for the Standardized input data
X_train_std, X_test_std = train_test_split(X_std, test_size=0.3, random_state=42)

# Inspecting the shape of the dataset after Standardization and taking a peak into it

print(f"\nNumber of samples and attributes of X after standardization: {X_std.shape}\n")
X_std.head()
```

```python
# 3) PCA Reduced Dataset

# Using PCA to reduce dimensionality of the data from 33 to 20, so that we take only the 15 most important features into consideration. 
# We take only 20, since a lot of them have a correlation even less than 0.05 with the target variable `price` (as seen in the correlation matrix). 

from sklearn.decomposition import PCA
pca_n = PCA(n_components=20)
X_pca = pd.DataFrame(pca_n.fit_transform(X_std))

X_train_pca, X_test_pca = train_test_split(X_pca, test_size=0.3, random_state=42)

print(f'Number of samples and attributes of X after PCA reduction: {X_pca.shape}')
X_pca.head()
```

```python
df_for_analysis = pd.concat([X_pca, y], axis=1)
```

#### Linear Regression

_Regular Data_

```python
# Creating a linear regression model and fitting it to the training data

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

# The model's score on the testing data

print("REGULAR DATA")
lin_score = model.score(X_test, y_test)
print('R2 score: ', lin_score)

# # Model's intercept
# print('\nIntercept: ', model.intercept_)

# # Model's coefficients
# print('\nCoefficients: ', model.coef_)
```
